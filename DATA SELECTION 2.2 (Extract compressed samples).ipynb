```python
import json
import os
from moviepy import VideoFileClip
```

```python
not_included_videos_dir = 'CADP/error-label-studio/'
skipped_videos_dir = 'label-studio-result/skipped/'
destination_directory = 'label-studio-result/'
```

```python
label_studio_output_filename = 'label-studio-annotations-2_attempt.json'
labels_filename = 'labels.json'
```

```python
with open(label_studio_output_filename, 'r') as file:
    label_studio_output = json.load(file)
```

```python
with open(destination_directory+labels_filename, 'r') as file:
    subclip_labels = json.load(file)
```

```python
for video in label_studio_output:
    file_upload = "-".join(video['file_upload'].split('-')[1:])
    video_id = video['id']

    annotations = video['annotations'][0]['result']
    subclip_id = 0
    try:
        clip = VideoFileClip(not_included_videos_dir+file_upload)
    except OSError as e:
        clip = VideoFileClip(skipped_videos_dir+file_upload)
        
    for subclip_annotations in annotations:
        subclip_filename = file_upload.split(".")[0] + f"_{subclip_id}.mp4"

        print(f'Cutting {subclip_filename}...')
        
        label = subclip_annotations['value']['labels'][0]
        subclip_labels[subclip_filename] = label

        start = subclip_annotations['value']['start']
        end = subclip_annotations['value']['end']
        subcli = clip.subclipped(start, end).without_audio()
        subcli.write_videofile(destination_directory+subclip_filename)

        print(f'{subclip_filename} saved...')
        
        subclip_id += 1
```

```python
with open(destination_directory+labels_filename, "w", encoding="utf-8") as file:
    json.dump(subclip_labels, file, ensure_ascii=False, indent=4)
```
