```python
import json
import os
from moviepy import VideoFileClip, concatenate_videoclips
```

```python
source_directory = 'CADP/for-label-studio/'
destination_directory = 'label-studio-result/'
skipped_directory = 'label-studio-result/skipped/'
```

```python
label_studio_output_filename = 'label-studio-annotations.json'
labels_filename = 'labels.json'
```

```python
with open(label_studio_output_filename, 'r') as file:
    label_studio_data = json.load(file)
```

```python
subclip_labels = {}
```

```python
for video in label_studio_data:
    file_upload = "-".join(video['file_upload'].split('-')[1:])
    
    video_id = video['id']
    annotations = video['annotations'][0]['result']
    indexed_annotations = list(enumerate(annotations))
    sorted_annotations_by_start = sorted(indexed_annotations, key=lambda t: t[1]['value']['start'])
    sorted_annotations_by_end = sorted(indexed_annotations, key=lambda t: t[1]['value']['end'])
    
    clip = VideoFileClip(source_directory+file_upload)

    if annotations == []:
        clip.write_videofile(skipped_directory+file_upload)
        continue
    
    by_start_id = 0
    by_end_id = 0
    
    while(by_start_id<len(sorted_annotations_by_start)):
        by_start_annotation = sorted_annotations_by_start[by_start_id][1]
        
        by_start_label = by_start_annotation['value']['labels'][0]
        by_start_start = by_start_annotation['value']['start']
        by_start_end = by_start_annotation['value']['end']

        by_end_annotation = sorted_annotations_by_end[by_end_id][1]
        
        by_end_label = by_end_annotation['value']['labels'][0]
        by_end_start = by_end_annotation['value']['start']
        by_end_end = by_end_annotation['value']['end']

        if by_start_start != by_end_start:
            subclip = clip.subclipped(by_end_start, by_end_end).without_audio()
            annotation_id = sorted_annotations_by_end[by_end_id][0]
            filename = file_upload.split(".")[0] + f"_{annotation_id}.mp4"
            subclip.write_videofile(destination_directory+filename)
            subclip_labels[filename] = by_end_label
            
            step1 = clip.subclipped(by_start_start, by_end_start-0.15).without_audio()
            step2 = clip.subclipped(by_end_end+0.15, by_start_end).without_audio()
            subclip = concatenate_videoclips([step1, step2])
            
            by_end_id+=1
        elif by_start_start == by_end_start and by_start_id != by_end_id:
            annotation_id = sorted_annotations_by_start[by_start_id][0]
            filename = file_upload.split(".")[0] + f"_{annotation_id}.mp4"
            subclip.write_videofile(destination_directory+filename)
            subclip_labels[filename] = by_start_label
            
            del step1
            del step2
            
            by_end_id+=1
            by_start_id = by_end_id
        elif by_start_start == by_end_start:
            annotation_id = sorted_annotations_by_start[by_start_id][0]
            filename = file_upload.split(".")[0] + f"_{annotation_id}.mp4"
            subclip = clip.subclipped(by_start_start, by_start_end).without_audio()
            subclip.write_videofile(destination_directory+filename)
            subclip_labels[filename] = by_start_label
            by_start_id+=1
            by_end_id+=1
```

```python
with open(destination_directory+labels_filename, "w", encoding="utf-8") as file:
    json.dump(subclip_labels, file, ensure_ascii=False, indent=4)
```
