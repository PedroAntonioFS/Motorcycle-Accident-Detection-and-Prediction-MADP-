```python
to_pop = []
for filename in committee.keys():
    for i in range(1,4):
        with open(f'peer-review-{filename}-{i}.json', 'r') as file:
            vote = json.load(file)
        
        for annotation in vote:
            file_upload = "-".join(annotation['file_upload'].split('-')[1:])
            file_upload = map_filename(file_upload)
            mapped_file_upload = '_'.join(file_upload.split("_")[:-1]) + ".mp4"
            if annotation['annotations'][0]['result'] and file_upload in input_labels_cleaned.keys():
                committee[filename][file_upload] = annotation['annotations'][0]['result'][0]['value']['choices'][0]
            elif annotation['annotations'][0]['result'] and mapped_file_upload in input_labels_cleaned.keys():
                committee[filename][file_upload] = annotation['annotations'][0]['result'][0]['value']['choices'][0]
                input_labels_cleaned[file_upload] = input_labels_cleaned[mapped_file_upload]
                to_pop.append(mapped_file_upload)
```

```python
for key in to_pop:
    input_labels_cleaned.pop(key, None)
```

```python
output_labels = {}
deadlock = {}
skipped = {}
for file_upload in input_labels_cleaned.keys():
    ballot_box = {
        'sem_acidente': 0,
        'acidente_sem_moto': 0,
        'acidente_com_moto': 0,
        'skipped': 0
    }
    
    for reviewer in committee.values():
        try:
            vote = reviewer[file_upload]
        except KeyError:
            vote = 'skipped'
        ballot_box[vote]+=1
        
    winner_qtd_votes = max(ballot_box.values())
    winners = [label for label, label_votes in ballot_box.items() if label_votes == winner_qtd_votes]
    minerva = input_labels_cleaned[file_upload]
    
    if len(winners) == 1 and winners[0]=='skipped':
        skipped[file_upload] = winners
    elif len(winners) == 1:
        output_labels[file_upload] = winners[0]
    elif minerva in winners:
        output_labels[file_upload] = minerva
    else:
        deadlock[file_upload] = winners
```

```python
frames_1fps = {}
tranformation_option1 = "frames-1fps/"
for filename in os.listdir(source_directory):
    if not filename.lower().endswith('.mp4'):
        continue
        
    clip = VideoFileClip(source_directory+filename)
    duration = int(clip.duration)
    
    label_map_filename = '_'.join(filename.split("_")[:-1]) + ".mp4"
    if filename in output_labels:
        label = output_labels[filename]
    else:
        label = output_labels[label_map_filename]
    for t in range(0, duration):
        output_filename = filename.split(".")[0] + f"_frame_{t:04d}.png"
        frame = clip.get_frame(t)
        img = Image.fromarray(frame)
        img.save(target_directory+tranformation_option1+label+'/'+output_filename)
        frames_1fps[output_filename] = label
```

```python
with open(target_directory+tranformation_option1+labels_filename, "w", encoding="utf-8") as file:
    json.dump(frames_1fps, file, ensure_ascii=False, indent=4)
```
